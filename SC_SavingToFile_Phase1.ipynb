```python
import pandas as pd
```

```python
dataset=pd.read_csv("insurance_pre.csv")
```

```python
dataset=pd.get_dummies(dataset,drop_first=True)
```

```python
independent=dataset[['age', 'bmi', 'children','sex_male', 'smoker_yes']]
dependent=dataset[['charges']]
```

```python
# Standard Scaler
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
independent=sc.fit_transform(independent)
```

```python
import pickle
filename="ScSaveToFile.sav"
pickle.dump(sc,open(filename,"wb"))
scaler=pickle.load(open("ScSaveToFile.sav","rb"))
```

```python
# model creation using GridSearchCV
from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVR
params_grid={"kernel":['rbf','poly','linear','sigmoid'],
             'C':[10,100,1000,2000,3000],
             'gamma':['auto','scale']}
grid=GridSearchCV(SVR(),params_grid,refit=True,verbose=3,n_jobs=-1)
grid.fit(independent,dependent)
```

```python
print(grid.best_params_)
re=grid.cv_results_
```

```python
grid_prediction = grid.predict(independent)
# r2 evaluation 
from sklearn.metrics import r2_score
r_score=r2_score(dependent,grid_prediction)
```

```python
print( "The Best parameter {}".format(grid.best_params_), " and R2 Score value{}".format(r_score))
```

```python
# save model 
import pickle
filename="PreprocessedInputFile.sav"
```

```python
pickle.dump(grid,open(filename,"wb"))
load_model=pickle.load(open("PreprocessedInputFile.sav","rb"))
```
